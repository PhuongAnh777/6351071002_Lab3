```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# C·∫•u h√¨nh hi·ªÉn th·ªã
plt.style.use('fivethirtyeight')
sns.set_palette("husl")
```

```python
# 1. NH·∫¨P D·ªÆ LI·ªÜU V√Ä TH·ªêNG K√ä C∆† B·∫¢N
print("Tr·∫ßn Ph∆∞∆°ng Anh - 6351071002")
print("\n1. NH·∫¨P D·ªÆ LI·ªÜU V√Ä TH·ªêNG K√ä C∆† B·∫¢N")
print("-" * 70)

try:
    data = pd.read_csv('Mall_Customers.csv')
    print("‚úì ƒê√£ ƒë·ªçc d·ªØ li·ªáu t·ª´ file Mall_Customers.csv")
except:
    print("‚ö† Kh√¥ng t√¨m th·∫•y file")

print(f"\nüìä K√≠ch th∆∞·ªõc d·ªØ li·ªáu: {data.shape[0]} h√†ng, {data.shape[1]} c·ªôt")
```

```python
# Hi·ªÉn th·ªã 5 d√≤ng ƒë·∫ßu
print("Tr·∫ßn Ph∆∞∆°ng Anh - 6351071002")
print("\nüìã Xem tr∆∞·ªõc d·ªØ li·ªáu (5 d√≤ng ƒë·∫ßu):")
print(data.head())

# Th√¥ng tin c∆° b·∫£n v·ªÅ d·ªØ li·ªáu
print("\nüìã Th√¥ng tin v·ªÅ d·ªØ li·ªáu:")
print(data.info())

# Th·ªëng k√™ m√¥ t·∫£
print("\nüìä TH·ªêNG K√ä M√î T·∫¢:")
print(data.describe())

# Ki·ªÉm tra gi√° tr·ªã thi·∫øu
print("\nüîç KI·ªÇM TRA D·ªÆ LI·ªÜU THI·∫æU:")
missing_data = data.isnull().sum()
print(missing_data)

if missing_data.sum() > 0:
    print(f"\n‚ö† T·ªïng s·ªë gi√° tr·ªã thi·∫øu: {missing_data.sum()}")
    print(f"‚ö† T·ª∑ l·ªá d·ªØ li·ªáu thi·∫øu: {(missing_data.sum() / (data.shape[0] * data.shape[1]) * 100):.2f}%")
else:
    print("\n‚úì Kh√¥ng c√≥ d·ªØ li·ªáu thi·∫øu!")

# X·ª¨ L√ù D·ªÆ LI·ªÜU THI·∫æU
print("\n" + "=" * 70)
print("X·ª¨ L√ù D·ªÆ LI·ªÜU THI·∫æU")
print("-" * 70)

if missing_data.sum() > 0:
    print("\nPh∆∞∆°ng ph√°p x·ª≠ l√Ω:")

    # X·ª≠ l√Ω Age: ƒëi·ªÅn b·∫±ng gi√° tr·ªã trung v·ªã
    if data['Age'].isnull().sum() > 0:
        age_median = data['Age'].median()
        data['Age'].fillna(age_median, inplace=True)
        print(f"  ‚Ä¢ Age: ƒêi·ªÅn {data['Age'].isnull().sum()} gi√° tr·ªã thi·∫øu b·∫±ng trung v·ªã ({age_median:.0f})")

    # X·ª≠ l√Ω Annual Income: ƒëi·ªÅn b·∫±ng gi√° tr·ªã trung b√¨nh
    if data['Annual Income (k$)'].isnull().sum() > 0:
        income_mean = data['Annual Income (k$)'].mean()
        data['Annual Income (k$)'].fillna(income_mean, inplace=True)
        print(f"  ‚Ä¢ Annual Income: ƒêi·ªÅn {data['Annual Income (k$)'].isnull().sum()} gi√° tr·ªã thi·∫øu b·∫±ng trung b√¨nh ({income_mean:.1f})")

    # X·ª≠ l√Ω Spending Score: ƒëi·ªÅn b·∫±ng gi√° tr·ªã trung b√¨nh
    if data['Spending Score (1-100)'].isnull().sum() > 0:
        spending_mean = data['Spending Score (1-100)'].mean()
        data['Spending Score (1-100)'].fillna(spending_mean, inplace=True)
        print(f"  ‚Ä¢ Spending Score: ƒêi·ªÅn {data['Spending Score (1-100)'].isnull().sum()} gi√° tr·ªã thi·∫øu b·∫±ng trung b√¨nh ({spending_mean:.1f})")

    print("\n‚úì ƒê√£ x·ª≠ l√Ω xong d·ªØ li·ªáu thi·∫øu!")
    print(f"‚úì S·ªë gi√° tr·ªã thi·∫øu c√≤n l·∫°i: {data.isnull().sum().sum()}")
else:
    print("‚úì Kh√¥ng c·∫ßn x·ª≠ l√Ω d·ªØ li·ªáu thi·∫øu")
```

```python
# Th·ªëng k√™ sau khi x·ª≠ l√Ω
print("Tr·∫ßn Ph∆∞∆°ng Anh - 6351071002")
print("\nüìä TH·ªêNG K√ä SAU KHI X·ª¨ L√ù:")
print("\nTh·ªëng k√™ theo gi·ªõi t√≠nh:")
print(data['Gender'].value_counts())
print(f"\nT·ª∑ l·ªá gi·ªõi t√≠nh:")
print(data['Gender'].value_counts(normalize=True) * 100)

print("\n" + "=" * 70)
print("TH·ªêNG K√ä CHI TI·∫æT C√ÅC THU·ªòC T√çNH")
print("-" * 70)

# Th·ªëng k√™ Age
print("\nüìä TU·ªîI (Age):")
print(f"  ‚Ä¢ Trung b√¨nh: {data['Age'].mean():.2f} tu·ªïi")
print(f"  ‚Ä¢ Trung v·ªã: {data['Age'].median():.2f} tu·ªïi")
print(f"  ‚Ä¢ ƒê·ªô l·ªách chu·∫©n: {data['Age'].std():.2f}")
print(f"  ‚Ä¢ Min - Max: {data['Age'].min():.0f} - {data['Age'].max():.0f} tu·ªïi")

# Th·ªëng k√™ Annual Income
print("\nüí∞ THU NH·∫¨P H·∫∞NG NƒÇM (Annual Income):")
print(f"  ‚Ä¢ Trung b√¨nh: ${data['Annual Income (k$)'].mean():.2f}k")
print(f"  ‚Ä¢ Trung v·ªã: ${data['Annual Income (k$)'].median():.2f}k")
print(f"  ‚Ä¢ ƒê·ªô l·ªách chu·∫©n: ${data['Annual Income (k$)'].std():.2f}k")
print(f"  ‚Ä¢ Min - Max: ${data['Annual Income (k$)'].min():.0f}k - ${data['Annual Income (k$)'].max():.0f}k")

# Th·ªëng k√™ Spending Score
print("\nüõí ƒêI·ªÇM MUA S·∫ÆM (Spending Score):")
print(f"  ‚Ä¢ Trung b√¨nh: {data['Spending Score (1-100)'].mean():.2f}")
print(f"  ‚Ä¢ Trung v·ªã: {data['Spending Score (1-100)'].median():.2f}")
print(f"  ‚Ä¢ ƒê·ªô l·ªách chu·∫©n: {data['Spending Score (1-100)'].std():.2f}")
print(f"  ‚Ä¢ Min - Max: {data['Spending Score (1-100)'].min():.0f} - {data['Spending Score (1-100)'].max():.0f}")
```

```python
print("Tr·∫ßn Ph∆∞∆°ng Anh - 6351071002")
print("\n" + "=" * 70)
print("2. BI·ªÇU ƒê·ªí PH√ÇN PH·ªêI (HISTOGRAM)")
print("-" * 70)
print("\n‚úì ƒêang v·∫Ω bi·ªÉu ƒë·ªì histogram cho c√°c thu·ªôc t√≠nh...")

plt.figure(figsize=(15, 6))
n = 0

# Danh s√°ch c√°c thu·ªôc t√≠nh c·∫ßn v·∫Ω
attributes = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']

for x in attributes:
    n += 1
    plt.subplot(1, 3, n)
    plt.subplots_adjust(hspace=0.5, wspace=0.5)

    # V·∫Ω histogram v·ªõi KDE (Kernel Density Estimation)
    sns.histplot(data[x], bins=20, kde=True, color='skyblue', edgecolor='black')

    # Th√™m ƒë∆∞·ªùng trung b√¨nh
    mean_val = data[x].mean()
    plt.axvline(mean_val, color='red', linestyle='--', linewidth=2,
                label=f'Mean: {mean_val:.1f}')

    # Th√™m ƒë∆∞·ªùng trung v·ªã
    median_val = data[x].median()
    plt.axvline(median_val, color='green', linestyle='--', linewidth=2,
                label=f'Median: {median_val:.1f}')

    plt.title(f'Distribution of {x}', fontsize=12, fontweight='bold')
    plt.xlabel(x, fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('mall_customer_histograms.png', dpi=300, bbox_inches='tight')
print("‚úì ƒê√£ l∆∞u bi·ªÉu ƒë·ªì: mall_customer_histograms.png")
plt.show()

# V·∫Ω th√™m boxplot ƒë·ªÉ ph√°t hi·ªán outliers
print("\n‚úì ƒêang v·∫Ω bi·ªÉu ƒë·ªì boxplot ƒë·ªÉ ph√°t hi·ªán outliers...")
plt.figure(figsize=(15, 5))

for i, col in enumerate(attributes, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=data[col], color='lightblue')
    plt.title(f'Boxplot of {col}', fontweight='bold')
    plt.ylabel(col)
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('mall_customer_boxplots.png', dpi=300, bbox_inches='tight')
print("‚úì ƒê√£ l∆∞u bi·ªÉu ƒë·ªì: mall_customer_boxplots.png")
plt.show()

# Ph√¢n t√≠ch ph√¢n ph·ªëi
print("\n" + "=" * 70)
print("PH√ÇN T√çCH PH√ÇN PH·ªêI D·ªÆ LI·ªÜU")
print("-" * 70)

from scipy import stats

for col in attributes:
    # Ki·ªÉm tra skewness (ƒë·ªô l·ªách)
    skewness = data[col].skew()
    print(f"\n{col}:")
    print(f"  ‚Ä¢ Skewness: {skewness:.3f}", end="")
    if abs(skewness) < 0.5:
        print(" (Ph√¢n ph·ªëi g·∫ßn nh∆∞ ƒë·ªëi x·ª©ng)")
    elif skewness > 0:
        print(" (Ph√¢n ph·ªëi l·ªách ph·∫£i)")
    else:
        print(" (Ph√¢n ph·ªëi l·ªách tr√°i)")

    # Ki·ªÉm tra kurtosis (ƒë·ªô nh·ªçn)
    kurt = data[col].kurtosis()
    print(f"  ‚Ä¢ Kurtosis: {kurt:.3f}", end="")
    if abs(kurt) < 0.5:
        print(" (Ph√¢n ph·ªëi g·∫ßn chu·∫©n)")
    elif kurt > 0:
        print(" (Ph√¢n ph·ªëi nh·ªçn h∆°n chu·∫©n)")
    else:
        print(" (Ph√¢n ph·ªëi b·∫πt h∆°n chu·∫©n)")

print("\n" + "=" * 70)
print("‚úì HO√ÄN TH√ÄNH PH√ÇN T√çCH!")
print("=" * 70)
```
