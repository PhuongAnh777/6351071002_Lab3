```python
# 1. In ra tên các thuộc tính và thuộc tính phân lớp
print("Trần Phương Anh - 6351071002")
print("\n1. TÊN CÁC THUỘC TÍNH:")
print("-" * 40)
for i, feature in enumerate(iris.feature_names, 1):
    print(f"   {i}. {feature}")

print("\n   THUỘC TÍNH PHÂN LỚP (Target):")
for i, target in enumerate(iris.target_names, 1):
    print(f"   {i}. {target}")
```
